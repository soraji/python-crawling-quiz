```python
import requests
import time
from selenium import webdriver
```

```python
from PIL import Image as pil
```

```python
url = "https://www.ted.com/talks/julius_maada_bio_a_vision_for_the_future_of_sierra_leone"
driver = webdriver.Chrome()
driver.get(url)

time.sleep(1)
driver.save_screenshot("screenshot.png")
driver.quit()
```

```python
down_img = pil.open('screenshot.png')
print(down_img.size)
down_img
```

```python
url = "https://www.ted.com/talks/julius_maada_bio_a_vision_for_the_future_of_sierra_leone"
driver = webdriver.Chrome()
driver.get(url)

element = driver.find_element_by_css_selector("#video-container")
location = element.location
size = element.size
print(location, size)
driver.quit()

left = location["x"]
top = location["y"]
right = left + size["width"]
bottom = top + size["height"]
area = (left, top, right, bottom)
print(area)


down_img = pil.open('screenshot.png')
down_img = down_img.crop(area)
down_img
```
